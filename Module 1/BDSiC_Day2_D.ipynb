#  A workflow for reading papers

Breaking down the steps of reading primary scientitif literature

From Jennifer Raff's blog post: [How to Read and Understand a Scientific Paper](https://violentmetaphors.com/2013/08/25/how-to-read-and-understand-a-scientific-paper-2/)

```python
import schemdraw
from schemdraw.flow import *
```

```python
with schemdraw.Drawing() as d:
	d.config(fontsize=10, unit=.5)
	d0 = Start(w = 4, h=1).label("0. Paper")
	d+= d0
	d+= Arrow()
	d1 = Process(w = 4, h=1).label("1. Read Intro").drop("E")
	d+= d1
	d+= Arrow().right(2)
	d+= Process(w = 4, h=1).label("2. Big Question").drop("S")
	d+= Arrow().down()
	d+= Process(w = 4, h=1).label("3. Background")
	d+= Arrow()
	d4= Process(w = 4, h=1).label("4. Specific Questions")
	d+= d4
	d+= Arrow()
	d+= Process(w = 4, h=1).label("5. Approach").drop("W")
	d+= Arrow().left(2)
	dm = Process(w = 4, h=1).label("6. Read Methods").drop("S")
	d+= dm
	d+= Arrow().down()
	d+= Process(w = 4, h=1).label("7. Read Results").drop("E")
	d+= Arrow().right(2)
	d8= Process(w = 4, h=1).label("8. Did the results answer \n the specific questions?").drop("S")
	d+= d8
	d+= Arrow(arrow='-').down(length = 1)
	d+= Arrow().left(length = 4)
	d9= Process(w = 4, h=1).label("9. Read Conslusion/Discussion").drop("S")
	d+= d9
	#d+= Wire('|-', k=10).at(d8.S)
	d+= Arrow().down()
	d+= Process(w = 4, h=1).label("10. Read Abstract").drop("E")
	d+= Arrow().right(2)
	d+= Process(w = 4, h=1).label("11. What do other resarchers say?").drop("S")
	d+= Arrow(arrow='-').down(length = 1)
	d+= Arrow().left(length = 4)
	d12= Process(w = 4, h=1).label("12. Browse References")
	d+= d12
	d+= Wire('c',k=-1, arrow="->").at(d12.W).to(d0.W)
	d+= Wire('c',k=1, arrow="->").at(d4.E).to(d8.E)

```

```python
d.draw()
```
