```python
import tarfile
import gzip
import os
import shutil
import pandas as pd
```

```python
# dictionary object that can store file contents in cataloged fashion
file_dict = {}
tar_names = ''

with tarfile.open('source_data/GSE146917_RAW.tar', 'r') as tar:
    tar_names = tar.getnames()
    for name in tar_names:
        tar.extract(name)
        if name.endswith(".csv.gz"):
            with gzip.open(name, 'rt') as f_in, open("source_data/StudyAugmented_platformManifest.csv", 'w') as f_out:
                shutil.copyfileobj(f_in, f_out)

        # Clean up extracted compressed file
        os.remove(name)
```
